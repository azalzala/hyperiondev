```python
reviews = df[['reviews.rating','reviews.text']].dropna()
reviews.shape
# Remove the columns where the name of the product isnt available 

```

```python
nlp = sp.load('en_core_web_sm')
def preprocess(text): 
    doc = nlp(text)
    return ' '.join([token.lemma_.lower() for token in doc if not token.is_stop and not token.is_punct])

reviews['processed_review'] = reviews['reviews.text'].apply(preprocess)
```

1. Review text is pre-processed by the NLP model using a function, which creates a new column in the dataset


Create function for sentiment analysis (taking review as input and output of sentiment)
1. Textblob assigns sentiment and polarity attributes to reviews, once stopwords have been removed 
2. Choose two products to compare similarity scores between two reviews

```python
def get_sentiment(review):
    blob = TextBlob(review) 
    sentiment = blob.sentiment.polarity
    return sentiment 

reviews['sentiment']= reviews['processed_review'].apply(get_sentiment)
```

```python
sentiment_values = reviews['sentiment'].values
for i, sentiment in enumerate(sentiment_values):
    if sentiment > 0: 
        reviews.loc[i, 'sentiment']= "positive"
    elif sentiment < 0: 
        reviews.loc[i, 'sentiment'] = "negative"
    else: 
        reviews.loc[i, 'sentiment'] = "neutral"
```
