```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv('/Users/ayazalzala/Downloads/balance.txt', sep=' ')
df.head()
```

Average Income based on ethnicity can be found using the .groupby(  ) method. The output should be the mean() value for each group. 

```python
getEthnicityincome= df.groupby('Ethnicity')['Income'].mean()
print(getEthnicityincome)
getEthnicityincome.plot(kind='bar', figsize=(5,4.5), ylabel="Income")
df.groupby('Ethnicity').count()
print(f"The average income of each ethnicity group ranges from 44 to 47 (rounded to 2sf). \n For African Americans, average income is {getEthnicityincome['African American']} (Sample size = 99). \n Caucasion average income is {getEthnicityincome['Caucasian']} (Sample size = 199). \n The Asian average income ranks last at {getEthnicityincome['Asian']} (Sample size = 102)")
```
